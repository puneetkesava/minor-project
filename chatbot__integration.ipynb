```python
%%writefile app.py
import streamlit as st
from transformers import pipeline, Conversation
from PIL import Image
import numpy as np
import tensorflow as tf

# Change this path to your mounted Drive path
MODEL_PATH = '/content/drive/MyDrive/smart_photo_organizer_model.h5'

classification_model = tf.keras.models.load_model(MODEL_PATH)

def is_duplicate(image):
    # Dummy duplicate detection - replace with your logic
    return False

def preprocess_image(image, target_size=(224,224)):
    image = image.resize(target_size)
    image_array = np.array(image) / 255.0
    image_array = np.expand_dims(image_array, axis=0)
    return image_array

def classify_image(image):
    img_array = preprocess_image(image)
    preds = classification_model.predict(img_array)
    class_idx = np.argmax(preds)
    class_names = ['selfie', 'passport', 'landscape', 'object']  # Change as per your classes
    return class_names[class_idx]

chatbot = pipeline("conversational", model="microsoft/DialoGPT-small")

st.title("AI Chatbot with Image Upload, Classification & Duplicate Detection")

user_input = st.text_input("You: ")

uploaded_file = st.file_uploader("Upload an image", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption='Uploaded Image', use_column_width=True)

    category = classify_image(image)
    st.write(f"Image classified as: **{category}**")

    duplicate = is_duplicate(image)
    if duplicate:
        st.warning("Duplicate image detected!")
    else:
        st.success("No duplicate detected.")

if user_input:
    conv = Conversation(user_input)
    response = chatbot(conv)
    st.text(f"Bot: {response.generated_responses[-1]}")
```
